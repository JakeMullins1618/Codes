Sometimes it is necessary to consider the second derivative when modeling a phenomenon.
Suppose that we have a mass lying on a flat, frictionless surface and that this mass is
attached to one end of a spring with the other end of the spring attached to a wall. We will
denote displacement of the spring by x. If x > 0, then the spring is stretched. If x < 0, the
spring is compressed. If x = 0, then the spring is in a state of equilibrium (Figure 1.1.4).
If we pull or push on the mass and release it, then the mass will oscillate back and forth
across the table.

By Newtons second law of motion, the force on the mass must be

oscillator. Harmonic oscillators are useful for modeling simple harmonic motion in mechanics.

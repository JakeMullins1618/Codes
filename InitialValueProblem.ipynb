```python
def matprint(mat, fmt="g"):
    col_maxes = [max([len(("{:"+fmt+"}").format(x)) for x in col]) for col in mat.T]
    for x in mat:
        for i, y in enumerate(x):
            print(("{:"+str(col_maxes[i])+fmt+"}").format(y), end="  ")
        print("")
```

# Matrix Version Damped Harmonic Oscillator

```python
model = np.array([[0,1],[-2,-0.1]])
```

```python
matprint(model)
```

```python
plt.matshow(model)
```

```python
def dx_dt(x,t):
    return model @ x 
```

```python
d = 2
x = np.array([0,1])

t  = 0
tf = 70
dt = 0.01

n = int((tf-t)/dt)
X = np.zeros((n,d))
```

```python
for i in range(n):

    x = x + dx_dt(x,t)*dt

    X[i] = x
```

```python
plt.plot(X[:, 0]);
```

# Matrix Version Predator-Prey System

```python
def dx_dt(x,t):
    return np.array([
                      alpha*x[0]  - beta*x[0]*x[1],
                                   delta*x[0]*x[1] - gamma*x[1]  
    ])
```

```python
d = 2
x = np.zeros(d,)

x[0] = 10
x[1] = 10

t  = 0
tf = 50
dt = 0.001

n = int((tf-t)/dt)
X = np.zeros((n,d))
```

```python
alpha = 1 
beta = 0.05 
delta = 0.02
gamma = 0.5
```

```python
for i in range(n):

    x = x + dx_dt(x,t)*dt

    X[i] = x
```

```python
plt.plot(X)
```

# Matrix Lorenz System

$\frac{dx}{dt}=\sigma(y-x)$

$\frac{dy}{dt}=x(p-z)-y$

$\frac{dz}{dt}=xy-\beta z$

```python
rho = 28.0
sigma = 10.0
beta = 8.0 / 3.0
```

```python
model = np.array([[-sigma,sigma,0,0,0],[rho,-1,0,0,-1],[0,0,-beta,1,0]])
```

```python
matprint(model)
```

```python
plt.matshow(model)
```

```python
def dx_dt(x, t):
    x, y, z = x
    state = np.array([x,y,z,x*y,x*z])
    return model @ state 
```

```python
d = 3
x = np.array([1,1,1])

t  = 0
tf = 50
dt = 0.001

n = int((tf-t)/dt)
X = np.zeros((n,d))
```

```python
for i in range(n):

    x = x + dx_dt(x,t)*dt

    X[i] = x
```

```python
fig = plt.figure()
ax = fig.gca(projection="3d")
ax.plot(X[:, 0], X[:, 1], X[:, 2]);
```
